```python
banjil_df = pd.read_csv(banjil_path)
banjil_df.head()
```

```python
# 1.   
plt.style.use('default')
plt.rcParams['figure.figsize'] = (7, 5)
plt.rcParams['font.size'] = 12

# 2.  
label_data = []
for i in range(2, label_df.shape[1]):
    label_data.append(label_df.iloc[:,i])

banjil_data = []
for i in range(2, banjil_df.shape[1]):
    banjil_data.append(banjil_df.iloc[:,i])

# 3.  
fig, ax = plt.subplots()

ax.violinplot(label_data)
ax.violinplot(banjil_data)
ax.set_ylim(0, 1)
ax.set_xlabel('blendshape')
ax.set_ylabel('Value')

plt.show()
```
