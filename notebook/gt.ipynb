```python
import os
import tarfile
import shutil
import pandas as pd
```

```python
gt_dir = '/mediazen/2022/2/raw/ok_takes/'
f_name_dir = '/shared/air/shared/youngkim/mediazen/csv_out/deepspeech_fc_finetune_expand_G_full_dataset-18:11:35:56/banjilbanjil'
```

```python
gt_names = []
for d in os.scandir(f_name_dir):
    gt_names.append("_".join(d.name.split("_")[1:4]))
```

```python
gt_paths = []
for d in os.scandir(gt_dir):
    if d.name in gt_names:
        gt_paths.append(os.path.join(d.path, f'{"_".join(d.name.split("_")[1:])}_mediazen_cal.csv'))
    
```
